```python
import pandas as pd
pd.set_option('display.max_rows', 60)

import matplotlib.pyplot as plt
```

```python
# Having a look at the data
df = pd.read_csv("linkedin_posts.csv", index_col=0)
df
```

```python
# Add helper column for analysis
df["text_len"] = df["texts"].str.len()

# I'm interested in reach of post (impressions)
eval_metric = "impressions"
df.sort_values(by=eval_metric, ascending=False).head()
```

```python
# Look at correlations
df[["reactions", "comments", "reposts", "impressions", "text_len"]].corr(method="pearson")
```

```python
# Simple plot, trying to get a grasp of optimal post lenght
plt.scatter(df["text_len"], df["impressions"])
plt.xlabel("Number of characters")
plt.ylabel("Impressions")
```

```python
plt.scatter(df["reactions"], df["impressions"])
plt.xlabel("Number of reactions")
plt.ylabel("Impressions")
```

```python
plt.scatter(df["comments"], df["impressions"])
plt.xlabel("Number of comments")
plt.ylabel("Impressions")
```

```python
plt.scatter(df["reactions"], df["impressions"])
plt.xlabel("Number of reactions")
plt.ylabel("Impressions")
```
